```python
name1=[]
mileage1=[]
dealername1=[]
rating1=[]
rating_count1=[]
price1=[]
for i in range(1,6):
    url1='https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=volvo&models%5B%5D=&list_price_max=&maximum_distance=all&zip='+str(i)
    
    page=requests.get(url1)
    page.status_code
    soup=BeautifulSoup(page.content,'html.parser')
    results=soup.find_all('div',{'class':'vehicle-card'})
    
    for result in results:
        try:
            name1.append(result.find('h2').get_text())
        except:
            name1.append(np.NAN)
        try:
            mileage1.append(result.find('div',{'class':'mileage'}).get_text())
        except:
            mileage1.append(np.NAN)
        try:
            dealername1.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
        except:
            dealername1.append(np.NAN)
        try:
            rating1.append(result.find('span',{'class':'sds-rating__count'}).get_text())
        except:
            rating1.append(np.NAN)  
        try:
            rating_count1.append(result.find('span',{'class':'sds-rating__link sds-button-link'}).get_text())
        except:
            rating_count1.append(np.NAN)
        try:
            price1.append(result.find('span',{'class':'primary-price'}).get_text())
        except:
            price1.append(np.NAN)
```

### Problem :
* Rating is in the Object we have changes into int

```python
cars['Rating Count']=cars['Rating Count'].apply(lambda x:x.strip('reviews)').strip('('))
cars['Rating Count'] = cars['Rating Count'].apply(lambda x:x.replace(",",""))
cars['Rating Count'] = cars['Rating Count'].astype(int, errors = 'raise')
cars
```
